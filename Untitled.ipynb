```python
class Population:
    def __init__(self,number, genotypes):
        self.number = number
        self.genotypes = genotypes
        self.phenotypes = {}
        
    def generate_phenotypes(self):
        for gene in self.genotypes:
            self.phenotypes[gene] = {'square': gene*gene, 'tenfold': gene*10}
        return self.phenotypes
```

```python
pop1 = Population(1, [2,4])
```

```python
pop1.generate_phenotypes()
```
