```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
sns.set_theme(style="whitegrid")
```

```python
lst_errorRates = [0.10650887573964496,
 0.03944773175542404,
 0.07297830374753456,
 0.04930966469428011,
 0.05522682445759364,
 0.10453648915187375]
lst_resnet = ['resnet18', 'resnet18', 'resnet18', 'resnet34', 'resnet34', 'resnet34']
lst_models_str = ['learn','learn_lr', 'learn_lr_fresh', 'learn', 'learn_lr', 'learn_lr_fresh']

df = pd.DataFrame(list(zip(lst_models_str, lst_resnet, lst_errorRates)), 
               columns =['Model','resnet', 'ErrorRate']) 
```

```python
df
```

```python
sns.set(font_scale=1.5)
g = sns.catplot(
    data=df, kind="bar",
    x="resnet", y="ErrorRate", hue="Model", palette="dark", alpha=0.7, height=6
)
g.despine(left=True)
g.set_axis_labels("", "Error rate (%)")
```
